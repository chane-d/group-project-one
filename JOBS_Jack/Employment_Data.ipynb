```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# get csv files

colorado_file = 'csv_folder/colorado.csv'
oregon_file = 'csv_folder/oregon.csv'
washington_file = 'csv_folder/washington.csv'
alaska_file = 'csv_folder/alaska.csv'
us_employment_file = 'csv_folder/us_nonfarm_ttl_employed.csv'
us_unemployment_rate_file = 'csv_folder/us_unemployment_rate.csv'
us_unemployed_file = 'csv_folder/us_unemployed_ttl.csv'
```

```python
# make dataframe

colorado_df = pd.read_csv(colorado_file)
oregon_df = pd.read_csv(oregon_file)
washington_df = pd.read_csv(washington_file)
alaska_df = pd.read_csv(alaska_file)
us_employment_df = pd.read_csv(us_employment_file)
us_unemployment_rate_df = pd.read_csv(us_unemployment_rate_file)
us_unemployed_df = pd.read_csv(us_unemployed_file)
```

```python
# US
```

```python
us_unemployment_rate_df.head()

us_output = []
us_year_list = us_unemployment_rate_df['Year'].unique()
us_year_list = us_year_list[3:]

for location in range(3,11):
    us_year = us_unemployment_rate_df.iloc[location]
    us_mean = round(us_year.iloc[1:].mean(), 2)
    us_output.append(us_mean)
    

us_ur_df = pd.DataFrame({
    'Year' : us_year_list,
    'US Unemployment Rate' : us_output
})
```

```python
# Colorado
```

```python
year_list = colorado_df['Year'].unique()
colorado_df = colorado_df[:-1]
colorado_df['unemployment rate'] = pd.to_numeric(colorado_df['unemployment rate'])

output = []

for year in year_list:
    by_year = colorado_df.loc[colorado_df['Year'] == year]
    mean_rate = round(by_year['unemployment rate'].mean(), 2)
    
    output.append(mean_rate)
    
output

amended_colorado_df = pd.DataFrame({
    'Year': year_list,
    'Colorado Unemployment Rate': output
})
```

```python
us_col = us_ur_df.merge(amended_colorado_df, on='Year')
us_col_diff = us_col['US Unemployment Rate'] - us_col['Colorado Unemployment Rate']
```

```python
%matplotlib notebook

colorado_year = amended_colorado_df['Year']
colorado_ur = amended_colorado_df['Colorado Unemployment Rate']

plt.plot(colorado_year, colorado_ur, marker ='o', color='palevioletred', label = 'Colorado')
plt.plot(us_year_list, us_output, marker ='o', color='dimgrey', label = 'US')

plt.legend()

plt.axvline(x = 2014, color = 'green')
plt.rcParams['axes.facecolor'] = 'whitesmoke'

plt.title('Yearly Mean Unemployment Rate (Colorado and Total US)')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
```

```python
plt.plot(us_year_list, us_col_diff, marker ='o', color='palevioletred')

plt.axvline(x = 2014, color = 'green')
plt.rcParams['axes.facecolor'] = 'whitesmoke'

plt.title('Difference Between US and Colorado Unemployment Rate by Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')


```
