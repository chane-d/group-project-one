```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
years = [2011,2012,2013,2014,2015,2016,2017,2018]
washington = [11.685013, -12.570129, 13.492889, 12.209658, -7.680784, 3.289571, 5.084262, 3.628703]
colorado = [2.809254, -8.771527, 16.310093, 7.011469, 0.748231, -8.189482, 1.661565, 4.021505]
oregon = [15.794251, -25.398484, 27.425569, 13.194063, -10.393395, -1.653117, 1.515569, 5.618893] 
usa = [8.000542, -3.953325, 14.360193, 8.332738, -0.879892, -0.523538, 1.603800, 3.140837]
```

```python


fig1,ax1=plt.subplots()
ax1.plot(years, colorado,color="palevioletred", label="Colorado",marker='o')
ax1.plot(years, usa, color="dimgrey", label="Total US",marker='o')

plt.legend(loc="best")
fig1.suptitle("Change in Tax Revenue (%) : Colorado vs US", fontsize=16, fontweight="bold")

plt.xlabel("Years")
plt.ylabel("Percent Change in Tax Revenue from Previous Year")
plt.xticks(years,rotation='vertical')
plt.axvline(x=2014,color='forestgreen')

ax1.set_facecolor('whitesmoke')

```

```python
fig2,ax2=plt.subplots()
ax2.plot(years, washington, color="mediumpurple", label="Washington",marker='o')

ax2.plot(years, usa, color="dimgrey", label="Total US",marker='o')
plt.legend(loc="best")

fig2.suptitle("Change in Tax Revenue (%) : Washington vs US", fontsize=16, fontweight="bold")

plt.xlabel("Years")
plt.ylabel("Percent Change in Tax Revenue from Previous Year")
plt.xticks(years,rotation='vertical')
plt.axvline(x=2014,color='forestgreen')

ax2.set_facecolor('whitesmoke')
```
