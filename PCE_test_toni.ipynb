```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pceraw=pd.read_csv('PCE_By_State.csv')
```

```python
pceraw
```

```python
pcedata=pceraw.loc[pceraw['Description']=='Personal consumption expenditures',:]
pcedata.reset_index(inplace=True)
pcedata=pcedata[['GeoName','ComponentName','Unit','Description','1997','1998','1999','2000','2001','2002','2003','2004','2005',
                 '2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017']]
```

```python
pcedata.head()
```

```python
pchange_1998=((pcedata['1998']-pcedata['1997'])/pcedata['1997'])*100
pchange_1999=((pcedata['1999']-pcedata['1998'])/pcedata['1998'])*100
pchange_2000=((pcedata['2000']-pcedata['1999'])/pcedata['1999'])*100
pchange_2001=((pcedata['2001']-pcedata['2000'])/pcedata['2000'])*100
pchange_2002=((pcedata['2002']-pcedata['2001'])/pcedata['2001'])*100
pchange_2003=((pcedata['2003']-pcedata['2002'])/pcedata['2002'])*100
pchange_2004=((pcedata['2004']-pcedata['2003'])/pcedata['2003'])*100
pchange_2005=((pcedata['2005']-pcedata['2004'])/pcedata['2004'])*100
pchange_2006=((pcedata['2006']-pcedata['2005'])/pcedata['2005'])*100
pchange_2007=((pcedata['2007']-pcedata['2006'])/pcedata['2006'])*100
pchange_2008=((pcedata['2008']-pcedata['2007'])/pcedata['2007'])*100
pchange_2009=((pcedata['2009']-pcedata['2008'])/pcedata['2008'])*100
pchange_2010=((pcedata['2010']-pcedata['2009'])/pcedata['2009'])*100
pchange_2011=((pcedata['2011']-pcedata['2010'])/pcedata['2010'])*100
pchange_2012=((pcedata['2012']-pcedata['2011'])/pcedata['2011'])*100
pchange_2013=((pcedata['2013']-pcedata['2012'])/pcedata['2012'])*100
pchange_2014=((pcedata['2014']-pcedata['2013'])/pcedata['2013'])*100
pchange_2015=((pcedata['2015']-pcedata['2014'])/pcedata['2014'])*100
pchange_2016=((pcedata['2016']-pcedata['2015'])/pcedata['2015'])*100
pchange_2017=((pcedata['2017']-pcedata['2016'])/pcedata['2016'])*100
```

```python
pchange = {
    "1998 Change": pchange_1998,
    "1999 Change": pchange_1999,
    "2000 Change": pchange_2000,
    "2001 Change": pchange_2001,
    "2002 Change": pchange_2002,
    "2003 Change": pchange_2003,
    "2004 Change": pchange_2004,
    "2005 Change": pchange_2005,
    "2006 Change": pchange_2006,
    "2007 Change": pchange_2007,
    "2008 Change": pchange_2008,
    "2009 Change": pchange_2009,
    "2010 Change": pchange_2010,
    "2011 Change": pchange_2011,
    "2012 Change": pchange_2012,
    "2013 Change": pchange_2013,
    "2014 Change": pchange_2014,
    "2015 Change": pchange_2015,
    "2016 Change": pchange_2016,
    "2017 Change": pchange_2017
}
pchange_df = pd.DataFrame(pchange)
pchange_df.head()
```

```python
countries=pcedata[['GeoName']]
all_change=countries.join(pchange_df)
all_change.head()
```

```python
CO=all_change.loc[all_change['GeoName']=='Colorado',['2000 Change','2001 Change','2002 Change',
                                                     '2003 Change','2004 Change','2005 Change','2006 Change','2007 Change',
                                                    '2008 Change','2009 Change','2010 Change','2011 Change','2012 Change',
                                                    '2013 Change','2014 Change','2015 Change','2016 Change','2017 Change']]
CO=CO.transpose()
```

```python
US=all_change.loc[all_change['GeoName']=='United States',['2000 Change','2001 Change','2002 Change',
                                                     '2003 Change','2004 Change','2005 Change','2006 Change','2007 Change',
                                                    '2008 Change','2009 Change','2010 Change','2011 Change','2012 Change',
                                                    '2013 Change','2014 Change','2015 Change','2016 Change','2017 Change']].transpose()
```

```python
# Create a list of the years that we will use as our x axis
years = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]

# Plot our line that will be used to track a wrestler's wins over the years
plt.plot(years, CO , color="green", label="Colorado")

# Plot our line that will be used to track a wrestler's losses over the years
plt.plot(years, US, color="blue", label="Losses")

# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")

plt.title("Colorado vs US PCE")
plt.xlabel("Years")
plt.ylabel("GDP (in millions of dollars)")
plt.xticks(years,rotation='vertical')

# Print our chart to the screen
plt.show()
```
