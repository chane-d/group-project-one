```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np
```

```python
pop_data_to_load = "../Inputs/POPSTATSInputs/population_estimates_project1.csv"

pop_data = pd.read_csv(pop_data_to_load)
pop_data
```

```python
#set index to Geography
pop_data = pop_data.set_index('Geography')
pop_data.head()
```

```python
new_pop_data = pd.DataFrame(columns=pop_data.columns)
new_pop_data
```

```python
new_pop_data['2010'] = pop_data['2010'].map(lambda x: int(x.replace(',', '')))
new_pop_data['2011'] = pop_data['2011'].map(lambda x: int(x.replace(',', '')))
new_pop_data['2012'] = pop_data['2012'].map(lambda x: int(x.replace(',', '')))
new_pop_data['2013'] = pop_data['2013'].map(lambda x: int(x.replace(',', '')))
new_pop_data['2014'] = pop_data['2014'].map(lambda x: int(x.replace(',', '')))
new_pop_data['2015'] = pop_data['2015'].map(lambda x: int(x.replace(',', '')))
new_pop_data['2016'] = pop_data['2016'].map(lambda x: int(x.replace(',', '')))
new_pop_data['2017'] = pop_data['2017'].map(lambda x: int(x.replace(',', '')))
new_pop_data['2018'] = pop_data['2018'].map(lambda x: int(x.replace(',', '')))
new_pop_data
```

```python
national_average = new_pop_data.iloc[1:52].mean()
national_average
```

```python
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
colorado = new_pop_data.loc['Colorado', ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']]
alaska = new_pop_data.loc['Alaska', ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']]
oregon = new_pop_data.loc['Oregon', ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']]
washington = new_pop_data.loc['Washington', ['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']]
```

```python
#create plots for legal states
legalyear1=new_pop_data.loc['Colorado']['2014']

fig1,ax1=plt.subplots()
ax1.plot(years, colorado, color="palevioletred", label="Colorado",marker='o')
ax1.plot(years, national_average, color="dimgrey", label="National Average",marker='o')

plt.legend(loc="best")
ttl=fig1.suptitle("Population Estimates 2010-2018:\n Colorado vs US National Average", fontsize=16, fontweight="bold")
ttl.set_position([.5,1])
plt.xlabel("Years")
plt.ylabel("Estimated Population (Millions)")
plt.xticks(years,rotation='vertical')
plt.axvline(x=2014,color='forestgreen')

ax1.set_facecolor('whitesmoke')

ax1.annotate('First Year of Legal Sales in Colorado: 2014', xy=(2014, legalyear1), xytext=(2015, 5800000),
            arrowprops=dict(facecolor='black', shrink=0.05))

plt.show()
```

```python
new_pop_data_transpose = new_pop_data.transpose()
new_pop_data_transpose
```

```python
pop_perc_change=new_pop_data_transpose.astype(float).pct_change()
pop_perc_change=pop_perc_change*100
pop_perc_change
```

```python
colorado_percent = pop_perc_change.loc[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018'], 'Colorado']
alaska_percent = pop_perc_change.loc[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018'], 'Alaska']
oregon_percent = pop_perc_change.loc[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018'], 'Oregon']
washington_percent = pop_perc_change.loc[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018'], 'Washington']
national_average_percent = pop_perc_change.mean(axis=1)


```

```python
fig8,ax1=plt.subplots()
ax1.plot(years, alaska_percent, color="lightseagreen", label="Alaska",marker='o')
ax1.plot(years, colorado_percent, color="palevioletred", label="Colorado",marker='o')
ax1.plot(years, oregon_percent, color="darkorange", label="Oregon",marker='o')
ax1.plot(years, washington_percent, color="mediumpurple", label="Washington",marker='o')
ax1.plot(years, national_average_percent, color="dimgrey", label="National Average",marker='o')

plt.legend(loc="best")
fig8.suptitle("Population Percent Change 2010-2018: Legal States vs US", fontsize=16, fontweight="bold")

plt.xlabel("Years")
plt.ylabel("Percent Change")
plt.xticks(years,rotation='vertical')


ax1.set_facecolor('whitesmoke')


plt.show()
```
